```
#q,qe,a,ae輸出成csv
with open(homePath + 'stc-3_emotion_train.json') as f:  
  train_data = json.load(f)
#print(train_data[0])



with open(homePath + 'stc-3_emotion_train.csv','w') as f:
  headers = ['question','question_emotion','answer','answer_emotion']
  f_csv = csv.writer(f)
  f_csv.writerow(headers)
  
  for i in range(len(train_data)):             #len(train_data)
    f_csv.writerow(train_data[i][0]+train_data[i][1])
    
```
