```

import json
import os

# Load the Drive helper and mount
from google.colab import drive

# This will prompt for authorization.
drive.mount('/content/drive')

homePath = '/content/drive/My Drive/Colab Notebooks/'
dataPath = homePath + 'data/answer/'
print(os.listdir(dataPath))
#!ls "/content/drive/My Drive/Colab Notebooks/"
```

```

with open(dataPath + 'test_dataset_post_only.json') as f:  #問句
  tqs = json.load(f)
with open(dataPath + 'r0.txt') as af:     #檢索式答案0
  rs = af.readlines()
with open(dataPath + 'g0.txt') as bf:     #生成式答案0
  gs0 = bf.readlines()
with open(dataPath + 'g1.txt') as cf:     #生成式答案1
  gs1 = cf.readlines()
with open(dataPath + 'g3_b204.txt') as df:     #生成式答案3_1
  gs31 = df.readlines()
with open(dataPath + 'g3_b128.txt') as ef:     #生成式答案3_2
  gs32 = ef.readlines()
with open(dataPath + 'g2.txt') as ff:     #生成式答案2
  gs2 = ff.readlines()
with open(dataPath + 'g4.txt') as gf:     #生成式答案4
  gs4 = gf.readlines()
print(len(tqs),len(rs),len(gs0),len(gs1),len(gs31),len(gs32),len(gs2),len(gs4))
```

```
import pandas as pd


pd.set_option('display.max_rows', 300)
pd.set_option('display.max_columns', 20)

pd_test = pd.DataFrame(index=range(len(tqs)), 
                        data={'question':tqs, 
                              'retrieval-answer':rs,
                              'generated-answer0':gs0,
                              'generated-answer1':gs1,
                              'generated-answer2':gs2,
                              'generated-answer3_1':gs31,
                              'generated-answer3_2':gs32,
                              'generated-answer4':gs4
                              
                              
                             
                             },
                       columns=['question','retrieval-answer','generated-answer0','generated-answer1','generated-answer2','generated-answer3_1','generated-answer3_2','generated-answer4'])
pd_test
```

```
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 300)
pd.set_option('display.max_columns', 20)


ans = pd.read_csv(dataPath + 'stc3_answer_v2.csv')
#ge0 = ans[['ge0']]
#ans2 = ans[ans.columns[1:]]
#print(type(ge0))
ans_array = np.array(ans)
print(ans_array[0][5])  #2
```

```

#ans_list = ans_array.tolist()
#print(ans_array[0])
result = []
#print(ans_array[0])
for i in range(200):
  post = tqs[i][0]
  emo = tqs[i][1]
  response0 = ans_array[i][2]
  responseemo0 = str(ans_array[i][3])
  response1 = ans_array[i][4]
  responseemo1 = str(ans_array[i][5])
  response2 = ans_array[i][6]
  responseemo2 = str(ans_array[i][7])
  response3 = ans_array[i][8]
  responseemo3 = str(ans_array[i][9])
  response4 = ans_array[i][10]
  responseemo4 = str(ans_array[i][11])
  result.append([[post,emo],[[response0,responseemo0],[response1,responseemo1],[response2,responseemo2],[response3,responseemo3],[response4,responseemo4]]])
#  print(post,emo,response0,responseemo0)
opfile = json.dumps(result,ensure_ascii=False)
#print(opfile)
with open(dataPath + 'op.json','w') as q:
  q.write(opfile)
```
